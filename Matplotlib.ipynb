```python
def f(t):
  return np.sin(2*np.pi*t)
t1=np.arange(0,5,0.1)
def f1(t):
  return np.exp(t)
t2=np.arange(0,5,0.02)
```

```python
plt.figure()
plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')

plt.subplot(212)
plt.plot(t2,f1(t2),'r--')
plt.show()
```
