```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Line Plot
x=[12,34,45,26,87,67,98]
y=[65,32,56,76,23,12,55]
sns.lineplot(x,y)
```

```python
#Scatter Plot
x=[12,34,45,26,87,67,98]
y=[65,32,56,76,23,12,55]
sns.scatterplot(x,y)
```

```python
#Bar Plot
data=pd.DataFrame(
    {
        'sub':['Hindi','English','Maths','Science'],
        'mark': [87,89,98,76]

    }

)
#data
sns.barplot(data=data,x='sub',y='mark')
```
